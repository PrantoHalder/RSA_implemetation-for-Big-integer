```python
import random
import math
_mrpt_num_trials = 5

def check_prime(n):

    assert n >= 2

    if n == 2:
        return True

    if n % 2 == 0:
        return False

    s = 0
    d = n - 1
    while True:
        quotient, remainder = divmod(d, 2)
        if remainder == 1:
            break
        s += 1
        d = quotient
    assert (2 ** s * d == n - 1)


    def try_composite(a):
        if pow(a, d, n) == 1:
            return False
        for i in range(s):
            if pow(a, 2 ** i * d, n) == n - 1:
                return False
        return True

    for i in range(_mrpt_num_trials):
        a = random.randrange(2, n)
        if try_composite(a):
            return False

    return True

def public_private_key(p, q):
    if p == q:
        raise ValueError('p and q are equal!')
    if not check_prime(p):
        p = find_next_prime(p)
    if not check_prime(q):
        q = find_next_prime(q)

    print('p: ', p)
    print('q: ', q)
    # n = pq
    n = p * q
    print('n: ', n)
    phi = (p - 1) * (q - 1)
    e = 3
    g = gcd(e, phi)
    while g != 1:
        e = e + 1
        g = gcd(e, phi)
    d = multiplicative_inverse(e, phi)
    print('e: ', e)
    print('d: ', d)
    print('phi: ', phi)

    return ((e, n), (d, n))

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a


def gcd_m(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = gcd_m(b % a, a)
        return (g, x - (b // a) * y, y)

def multiplicative_inverse(a, m):
    g, x, y = gcd_m(a, m)
    if g != 1:
        raise Exception('Modular Inverse Does Not Exist')
    else:
        return x % m

def find_next_prime(number):
    """Return the next prime."""
    next_number = number + 1
    while not check_prime(next_number):
        next_number += 1
    return next_number


def encryption(m, public_key, private_key):

    e, n = public_key

    d, n = private_key

    r = 2

    g = gcd(r, n)
    while g != 1:
        r = r+1
        g = gcd(r, n)


    ciphertext = pow(m,e,n)
    print ('RSA Encryption: Ciphertext: ', ciphertext)
    
    plaintext = pow(ciphertext,d,n)
    print('RSA Decryption: Plaintext:', plaintext)




def decrypt(pk, ciphertext):

    key, n = pk

    plain = [chr((char ** key) % n) for char in ciphertext]

    return ''.join(plain)


if __name__ == '__main__':

    print('Enter Large Prime Number, P: ')
    p = int(input())
    #print(p)
    print('Enter Another Large Prime Number, Q: ')
    q = int(input())
    #print(q)
    public_key, private_key = public_private_key(p, q)
    print("Public key is ", public_key)
    print("Private key is ", private_key)
    print('Enter Message, M: ')
    m = int(input())
    print(m)
    encryption(m, public_key, private_key)
```
